```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("homeprices (2).csv")
```

```python
print(df)
```

```python
model = LinearRegression()
model.fit(df[['area']], df['price'])
prediction = model.predict(pd.DataFrame([[1800]], columns=['area']))
print("predicted price:",prediction)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['area'], df['price'], color='red')
plt.plot(df['area'], model.predict(df[['area']]), color='blue')
plt.xlabel("Area")
plt.ylabel("price")
plt.title("House price prediction")
plt.show()
```

```python
area_input = int(input("Enter area:"))
prediction = model.predict([[area_input]])
print("predicted price:",prediction)
```

```python
area_input = int(input("Enter area in sq ft:"))
prediction = model.predict([[area_input]])
print(f"Estimated price for {area_input} sq ft area is ${int(prediction[0])}")
```
